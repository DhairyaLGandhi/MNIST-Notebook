## The Model

This is a fairly straightforward MLP that will yield okay-ish results on MNIST, nothing too great. There are a multitude of techniques that people have come up with to learn the dataset better. 

```julia
model = m = Chain(
  Dense(dims[1]^2, 32, relu),
  Dense(32, 10),
  softmax)

# Note that Julia is 1 indexed
```

A `Chain` is used to define an abstracted model in Flux. It is a regular function that takes as many functions as you want as its arguments, and will propagate any input through it. The cool thing about the way its defined is that its just a regular Julia function, as are all the layers that we used inside it.

```julia
# Tiny Demo:

demo_model = Chain(x -> x^2, 
                   x -> x + 5)

demo_model(4)
```

As you can see, any arbitrary julia functions can be used inside of `Chain` to extend its usability way beyond simple ML.

To wit: <br>
    $input = 4$ <br>
    $4^2 = 16$ <br>
    $16 + 5 = 21$

We now have a model!
But there's one issue still. How do we know _how_ it will train.

```julia
loss(x, y) = crossentropy(m(x), y)
```

Great, now our little model can use the loss function to know what it needs to do. Don't they grow up so fast? :')

There is another issue, we need to know how well the model is performing. We need to define a metric to track.

```julia
accuracy(x, y) = mean(argmax(m(x)) .== argmax(y))
```
